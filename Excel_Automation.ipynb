```python
from openpyxl import load_workbook
```

```python
# Load workbook
wb = load_workbook("/content/test.xlsx")
```

```python
ws_final = wb["final"]
ws_data = wb["data"]
ws_pvt = wb["PVT"]
```

```python
# --- Count non-empty cells in column A ---
def count_non_empty(ws):
    return sum(1 for cell in ws["A"] if cell.value not in (None, ""))
```

```python
# --- Compute N and T ---
N = max(0, count_non_empty(ws_data) - 1)
T = max(0, count_non_empty(ws_pvt) - 1)
```

```python
# --- Total rows to keep after row 16 ---
rows_to_keep = N + T
```

```python
start_row = 17
last_row_to_keep = start_row + rows_to_keep - 1
```

```python
# --- Total rows currently in sheet ---
max_row = ws_final.max_row
```

```python
# --- Delete extra rows ---
if max_row > last_row_to_keep:
    ws_final.delete_rows(last_row_to_keep + 1, max_row - last_row_to_keep)
```

```python
# Save
wb.save("updated.xlsx")
```
